# ðŸ”¹Booleans (True and False)
 - ###### Logical and comparison operators make boolean expressions
 - ###### Boolean expressions used in control statements
 - ###### Chaining boolean operators in one expression (read "and")



```python
# Booleans and Logical Operators
not True  # => False
True and False  # => False
True or False  # => True. Short-circuits

# Booleans and Comparison Operators
1 == 1  # => True
2 * 3 == 5  # => False
1 != 1  # => False
2 * 3 != 5  # => True

1 < 10  # True
2 >= 0  # True

# <<advanced>> Chained Comparison
1 < 2 < 3  # True. Computes 1 < 2 and 2 < 3

# Booleans and conditional control statements (if/else)
x = 10.6
if x > 0 and x <=10 :
    print ("positive number equal or less than 10")
elif x > 0:
    print ("positive number greater than 10")
else:
    print ("Negative Number")
```

# ðŸ”¹Numerics  (int and float)
- ###### int is a whole number while float is for floating decimal point number
- ###### math operators make numeric expressions
- ###### float is contagious in numeric expressions
- ###### division (float expression) vs. floor division (int expression)
- ###### division vs. modulus vs. remainder operators
- ###### ** is the exponentiation operator


```python
# int is a whole number while float is for floating decimal point number
3  # => 3 (int)
3.0  # => 3.0 (float)

# math operators make numeric expressions
# float is contagious in numeric expressions
1 + 1    # => 2 (int)
1.0 + 1  # => 2.0 (float)
8 - 1    # => 7 (int)
10 * 2   # => 20 (int)
8 * 2.5  # => 20.0 (float)
5 / 2    # => 2.5 (float)
13 / 4   # => 3.25 (float)
9 / 3    # => 3.0 (float)
7 / 1.4  # => 5.0 (float)

# division (float expression) vs. floor division (int expression)
# division vs. modulus vs. remainder operators
# ** is the exponentiation operator
7 // 3 # => 2 (int division)
7 % 3  # => 1 (int modulus)
3 ** 2 # => 9 (exponentiate)
```

# ðŸ”¹Strings (str)
- ###### Python has **no character type** (only str)
- ###### Both **single and double quotes** create string literals
- ###### Python also supports **Unicode** by default
- ###### Strings can be **concatenated** with +:
- ###### String **formatting** can be done using f function, curly brackets place holders and expressions
- ###### **Multiline** string literals can be introduced and closed with three single quotes or three double quotes
- ###### Multiline string literals are often used as documentation strings for modules, classes, and functions that you'll write
- ###### Special characters in string literals can be **escaped** with the \ character. 
- ###### Choose the string delimiter that makes your text easy to work with.
- ###### To avoid escaping, you can prefix the string literal with an r, like r"U\nescaped", whose second character is '\'.
- ###### There are many **string methods** we can use, for that use python documentation to look them up



```python
# Python has no character type (only str)
# both ' and " create string literals
# Python also supports Unicode by default
greeting = 'Hello'
group = "wÃ¸rld"

# Strings can be concatenated with +:
greeting + ' ' + group + "!" # => 'Hello wÃ¸rld!'

# String formatting can be done using f function, curly brackets place holders and expressions
f"{greeting} {group}!"  # f-strings â€“ interpolate expressions into a string

# Other string formatting styles:
"{greeting} {group}!".format(greeting=greeting, group=group)  # string-formatting by named groups
"{0} {1}!".format(greeting, group)  # string-formatting by positional groups
"%s %s!" % (greeting, group)  # C-style string-formatting

# Multiline Strings
'''A very very
very very
very long string.
'''
print ('''A very very
very very
very long string.
''')

# Escaping Special Characters
print("this is a \n two line sentence") # => prints two lines
print(r"this is a \n one line sentence") # => prints one lines
```

```python
# String Methods
greeting = "Hello world! "

greeting.find('lo')   # 3 (-1 if not found)
greeting.replace('llo', 'y')  # => "Hey world!"
greeting.startswith('Hell')  # => True
greeting.isalpha()           # => False (due to '!')

greeting.lower()  # => "hello world! "
greeting.title()  # => "Hello World! "
greeting.upper()  # => "HELLO WORLD! "

greeting.strip()  # => "Hello world!"
greeting.strip('dH !')  # => "ello worl"
```

# ðŸ”¹Type Conversion and "Truthiness"
- ###### The named types str, int, float, and bool can be used as functions to convert values from one type to another.
- ###### Truthiness: Every Python object is implicitly convertible to a boolean value.
- ###### Empty objects are "Falsy" (i.e., False), all other objects are "truth" (i.e., True)



```python
# The named types str, int, float, and bool can be used as functions to convert values from one type to another.
str(42)       # => "42"
int("42")     # => 42
float("2.5")  # => 2.5
float("1")    # => 1.0
int("42.5")   # Raises an error (ValueError).

# Truthiness: Every Python object is implicitly convertible to a boolean value.

# Empty objects are "Falsy" (i.e., False)
bool(None)   # => False
bool(False)  # => False
bool(0)      # => False
bool(0.0)    # => False
bool('')     # => False
# Empty data structures too!
bool([])     # => False

# all other objects are "truth" (i.e., True)
bool(41)     # => True
bool('abc')  # => True
bool(int)    # => True
# Non-empty containers too!
bool([False])  # => True
```
