# ðŸ”¹Getters and Setters
 -  The @property (property decorator) makes a method as a getter so we can just access it as object property (not method)
 -  Another decorator (see below) is used to make a setter method

```python
class Distance:
    _KILOMETERS_PER_MILE = 1.60934

    def __init__(self, km):
        self.km = km

    @property
    def miles(self):
        return self.km / self._KILOMETERS_PER_MILE

    @miles.setter
    def miles(self, miles):
        self.km = miles * self._KILOMETERS_PER_MILE

dist = Distance(10)

# accessing a 
print(dist.miles)

# setting the distance to 50 miles
dist.miles = 50

print(dist.km)
print(dist.miles)
```
