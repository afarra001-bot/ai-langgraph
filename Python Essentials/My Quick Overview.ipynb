# ðŸ”¹Data Structures

## Lists

```python
# Lists are list of objects (anything) or sublists, like Collections in Java
# Sorted, Not Unique, Read/Write

my_list = [1, 2, "Hi", [], [7,8,9]]
my_list.append("end of the list")
print (my_list)
len(my_list)
```

## Sets

```python
# A Set is a group of unique elements
# Not sorted, Unique, Read/Write, 

my_set = {1,2,3,4,5}
print(my_set)
len(my_set)
type(my_set)
```

## Tuples

```python
# Tuples are mutable lists
# Sorted, Not Unique, Read Only

my_tuple = (1,2,3)
len(my_tuple)
(1,2) == (2,1)
```

## Dictionaries

```python
# Dicitionaries are like Sets, but they are key/value pairs
# Not Sorted, Keys should be Unique, Read/Write

my_dictionary = {
    'apple': 'A red fruit', 
    'bear': 'A scary animal',
    'apple': 'Sometimes a green fruit'
}

my_dictionary['apple'] # it takes the last key/value pair defind for that key
```

# ðŸ”¹Operators

## Arithmetic Operators

```python
x = 5
y = 2
Print(x + y)
Print(x * y)
Print(x ** y) # expo
Print(x / y)
Print(x % y) # remainder
```

## Comparison, Logical, and Membership Operators

```python
# Comparison Operators
x = 5
y = 2
Print(x < y)
Print(x > y)
Print(x <= y)
Print(x >= y)

# Logical Operators (and, or, not)
print(True and False) # False
print(True and True) # True
print(False and False) # False
print(True or False) # True
print(True or True) # True
print(False or False) # False
print(not True) # False


# Membership Operators
print (1 in [1,2,3,4,5])
print (10 not in [1,2,3,4,5])
```

# ðŸ”¹Control Flow

## If / Else statements

```python
# Notice the indentation
a = True
if a:
    print('It is true!')
    print('Also print this')
else:
    print('It is false!')
print('Always print this')
```

```python
# Indentation is critical in if/else blocks and nesting if
a = True
b = False
c = True
if a:
    print('It is true!')
    print('Also print this')
    if b:
        print('Both are true')
        if c:
            print('All three are true')
else:
    print('It is false!')
print('Always print this')
```

## For loops

```python
# for loops over iterator
# Notice the "in" here is not the same as in the "Membership" logical operators

a = [1,2,3,4,5]
for number in a:
    print(number)

```

## While loops

```python
# while takes a logical condition

a = 0
while a < 5:
    print(a)
    a = a + 1
```

# ðŸ”¹Functions

```python
# Naming convention is the same as variables (star with lowercase letters)
# pass value vs. pass reference


def multiply(val1, val2):
    return val1 * val2

def appendNumber(myList, num):
    myList.append(num)

def incrementValue(val):
    val = val + 1

# object types are passed by reference
a = [1,2,3]
appendNumber(a, 4)
print(a) # [1,2,3,4]

# primitve types are passed by value
x1 = 5
incrementValue(x1)
print(x1) # 5 (not 6)
```

# ðŸ”¹Classes & Objects

```python
# More in-depth needed for OOP in Python

# Naming conventions: Class starts with upper case, object variables with lower case
# the use of "__init__" and "self"
# Defining the class attributes inside the init function
# init function can take more parameters than just self

class Dog:
    def __init__(self, name):
        self.name = name
        self.legs = 4
    
    def speak(self):
        print(self.name + ' says: Bark!')


my_dog = Dog('Rover')
another_dog = Dog('Fluffy')
my_dog.speak()
another_dog.speak()
```
