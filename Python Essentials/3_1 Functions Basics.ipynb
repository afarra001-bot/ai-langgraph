# ðŸ”¹Returns
- Python Functions always return an object reference (no return means None object)
- No return can be done in three ways: return, pass or no return statement at all
- Multiple value retun is possible through tuple packing & unpacking mechanism

```python
# No return:
def no_return():
    print('function executed')
    return
    #pass

x = no_return()
print(x)
   

# Mulitple return
def multiple_return():
    print('function executed')
    return 5, 6, 7
    
x = multiple_return()
print(x)

x, y, z = multiple_return()
print(x)
```

# ðŸ”¹Function Scope vs. Global Scope

```python
# Local vs. global namespaces

x = 2 # global scope                                                                                                                                                                                        
def fun(y):
    x = 41 # local scope, doesnt override global scope
    z = 5 # local scope
    y = 33 # local scope 
    print(locals())  # A 
    print(globals()['x'])  # B 
    print(x, y, z)  # C 

fun(3)
```

# ðŸ”¹Pass by Object Reference (not value)
- Python is consistent: it always handles objects via references.


```python
# Test the object reference

def fun(y):
    y = 33 
    print(y)

rr = 22
fun(rr)
print(rr)

def fun2(lst):
    lst.append(4) 
    print(lst)

lst2 = [4,5,7]
fun2(lst2)
print(lst2)
```
